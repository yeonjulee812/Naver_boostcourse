## 분석과 시각화를 위한 tiny data 만들기

```python
df= df_raw.melt(id_vars=['국가(대륙)별','상품군별','판매유형별'], 
                var_name='기간', value_name='백만원')
df.shape
```

```python
df.head()
```

```python
df.info()
```

## 데이터 전처리 

### 기간에서 연도, 분기를 분리하기

```python
int('2019 4/4 p)'.split()[0])
```

```python
df['연도'] = df['기간'].map(lambda x: int(x.split('.')[0]))
df.head()
```

```python
df['분기'] = df['기간'].map(lambda x: int(x.split('.')[1].split('/')[0]))
df.head()
```

### 금액을 수치데이터로 표현하기 위해 데이터 타입 변경하기

```python
import numpy as np

df['백만원'] = df['백만원'].replace('-', np.nan).astype(float)
df['백만원']
df.head()
```

### 필요없는 데이터 제거하기

```python
df.info()
```

```python
#합계 데이터는 따로 구할 수 있기 때문에 제거
df = df[(df['국가(대륙)별']!='합계')&(df['상품군별']!='합계')].copy()
```

```python
df.info()
```

```python
df.isnull().sum()
```

## K-Beauty 시각화
### 전체 상품군 판매액

```python
df_total = df[df['판매유형별']=='계'].copy()
df_total.head()
```

```python
sns.lineplot(data=df_total, x='연도', y='백만원')
```

```python
sns.lineplot(data=df_total, x='연도', y='백만원', hue='상품군별')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) # 범례를 바깥으로
#화장품이 판매량 크게 증가
```

```python
sns.relplot(data=df_total, x='연도', y='백만원', 
            hue='상품군별', kind='line', col='상품군별', col_wrap=4)
# 다른 상품군별에 비해 화장품이 특히 크게 증가함
# 그러나 화장품 이외의 상품군의 증가세는 뚜렷하게 나타나지 않는 문제
```

```python
df_sub = df_total[-df_total['상품군별'].isin(['화장품'])].copy()
```

```python
sns.relplot(data=df_sub, x='연도', y='백만원', 
            hue='상품군별', col='상품군별', col_wrap=4, kind='line')
# 화장품을 제외하고는, 의류 및 패션 관련상품의 증가세가 두드러짐.
```

```python
df_sub = df_total[-df_total['상품군별'].isin(['화장품', '의류 및 패션 관련상품'])].copy()
```

```python
sns.relplot(data=df_sub, x='연도', y='백만원', 
            hue='상품군별', col='상품군별', col_wrap=4, kind='line')
#이외에는 가전전자통신기기, 음반 비디오 등 크게 증가
```

### 화장품의 온라인쇼핑 해외직접판매액

```python
df_cosmetic = df_total[df_total['상품군별']=='화장품'].copy()
df_cosmetic['상품군별'].unique()
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=df_cosmetic, x='연도', y='백만원', hue='분기')
# 1분기에서 4분기로 갈수록 판매액이 증가하는 것은 시간에 따라 점점 성장하는 추세이기 때문
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x='기간', y='백만원')
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic, x='기간', y='백만원', hue='국가(대륙)별')
# 중국의 판매액이 높다
# 중국 이외의 국가들에 대한 판매액이 잘 안보이는 문제
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_cosmetic[df_cosmetic['국가(대륙)별']!='중국'],
             x='기간', y='백만원', hue='국가(대륙)별')
# 아세안,기타, 미국 순으로 판매액 증가
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[df['판매유형별']!='계'].copy()
sns.lineplot(data=df_sub, x='기간', y='백만원', hue='판매유형별')
# 판매유형별 데이터는 2017년부터 있음
# (온라인) 면세점이 성장하고 있음
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
df_sub = df[(df['판매유형별']!='계') &(df['판매유형별']!='면세점')].copy()
sns.lineplot(data=df_sub, x='기간', y='백만원', hue='판매유형별', ci=None)
# 판매유형별 데이터는 2017년부터 있음
# 면세점 이외의 경우 전반적으로 성장하는 추세
```

### 의류 및 패션관련 상품 온라인쇼핑 해외직접판매액

```python
df_fashion = df[(df['상품군별']=='의류 및 패션 관련상품') & (df['판매유형별']=='계')].copy()
df_fashion.head()
# '계'는 2014년부터, 이외의 경우 2017년부터 데이터 존재
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x='기간', y='백만원', hue='국가(대륙)별')
# 판매액 순위는 중국 일본 미국 아세안 순
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x='기간', y='백만원', hue='판매유형별')
```

```python
df_fashion2 = df[(df['상품군별']=='의류 및 패션 관련상품') & (df['판매유형별']!='계')].copy()
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2, x='기간', y='백만원', hue='판매유형별', ci=None)
# 면세점 이외가 보통 판매액이 많았으나, 면세점 판매액의 증가로 이제는 판매액이 비슷해짐.
```
